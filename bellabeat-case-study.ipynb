### **Comparing Weight Loss of FitBit Users**

I first want to show that tracking health data leads to wieght loss results. The "Fitbit Fitness Tracker Data" dataset contains weight logs of the Fitbit users over two months. First I cleaned the FitBit Fitness Tracker Data in SQL. I joined the weight log tables of two seperate months of data (listed above) to each other by common user Ids. I used Inner Join because I only want those Ids that exist in both tables. I selected the maximum difference in wieght over those two months to draw my comparison. I grouped by user Id because some users logged multiple weigh-ins in a single month. Then I used the "row_number" function to assign the perviously seven digit user ids to a neater looking sequential format.




I imported that table into Kaggle to create a plot using R code (see above and below).

```r
ggplot(fitbitwl, 
       aes(x = user_id, 
           y = pounds_lost, 
           fill = pounds_lost > 0)) + 
  ggtitle("Number of Pounds Lost per FitBit User") + 
  geom_bar(stat = "identity") + 
  scale_fill_manual(guide = FALSE,
                    values = c("red", "green")) +
  theme(text=element_text(size=20))
```

*Note: There is a potential outlier here. User 2 gained about 8.5 pounds instead of losing weight. It could be due to an input error but I do not know for certain so I decided to keep the outlier in place.*

This data shows that 5/6 (83%) of poeple who use Smart technology see results in weight loss only after one month!
